```python
#-------------- ANZAHL AUFRUFE AM TAG X IM DEZEMBER

def sort_per_day(csv_file, day):
    lines_for_day = []
    for i in range(len(csv_file)):
        line = csv_file.iloc[i]
        #print(line["Datum"][:2])
        if line["Datum"][:2] == str(day):
            #print(line["Datum"]) 
            lines_for_day.append(line)
        elif line["Datum"][:2] == "0" + str(day):
            #print(line["Datum"])
            lines_for_day.append(line)
    #print(f"Liste mit Linien für Tag {day}: {lines_for_day}")
    return lines_for_day
#sort_per_day(14)


# die werte für x und y -achse definieren
x_achse = [i for i in range(1, 25)]
#print(x_achse[0])

y_achse = []
for i in range(24):
    y_achse.append(len(sort_per_day(data_dez, x_achse[i])))

print(f"X-Achse: {x_achse}")
print(f"Y-Achse: {y_achse}")


# diagramm zeichen mit x und y-achse
sns.barplot(x=x_achse, y=y_achse)

# Diagramm-Anpassungen
plt.title('Anzahl Aufrufe des Adventskalenders bis zum 24. Dezember')
plt.xlabel('Dezembertag')
plt.ylabel('Anzahl Aufrufe am Tag x')

plt.savefig('Diagramme/Aufrufe_dez.png')

# Diagramm anzeigen
plt.show()
```

```python
#-------------- ANZAHL AUFRUFE AM TAG X IM DEZEMBER UND JANUAR

# die werte für x und y -achse definieren
x_achse_dez = [i for i in range(1, 25)]
#print(x_achse[0])

x_achse_jan = [i for i in range(1, 12)]

x_achse_dez_jan = x_achse_dez + x_achse_jan

y_achse_dez = []
for i in range(len(x_achse_dez)):
    y_achse_dez.append(len(sort_per_day(data_dez, x_achse_dez[i])))

y_achse_jan = []
for i in range(len(x_achse_jan)):
    y_achse_jan.append(len(sort_per_day(data_jan, x_achse_jan[i])))

y_achse_dez_jan = y_achse_dez + y_achse_jan

x_achse_dez = [f"Dez-{i}" for i in range(1, 25)]
x_achse_jan = [f"Jan-{i}" for i in range(1, 12)]
x_achse_dez_jan = x_achse_dez + x_achse_jan

print(f"X-Achse: {x_achse_dez_jan}")
#print(f"Y-Achse im Dezember: {y_achse_dez}")
#print(f"Y-Achse im Januar: {y_achse_jan}")
print(f"Y-Achse Dez/Jan: {y_achse_jan}")

# diagramm zeichen mit x und y-achse
sns.barplot(x=x_achse_dez_jan, y=y_achse_dez_jan)

# Diagramm-Anpassungen
plt.title('Anzahl Aufrufe des Adventskalenders vom 1.Dezember bis 11.Januar')
plt.xlabel('Datum')
plt.ylabel('Anzahl Aufrufe am Tag x')
plt.xticks(rotation=90)  # X-Achsen-Beschriftungen drehen, damit sie besser lesbar sind

plt.savefig('Diagramme/Aufrufe_dez_jan.png')

# Diagramm anzeigen
plt.show()
```
