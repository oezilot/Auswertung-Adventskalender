```python
#-------------- Importe etc.

#Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns

#Getting the data in pandas DataFrame Format (jede zeile ist ein aufrufder website!)
data = pd.read_csv('../Data/filtered_advent_log.csv')
#print(data.head()) #Printing first 5 values in the dataset

'''
data_dez = pd.read_csv('../Data/filtered_advent_log_dez.csv')

data_jan = pd.read_csv('/home/zoe/Projects/Auswertung-Adventskalender/Data/filtered_advent_log_jan.csv')
'''
```

```python
#-------------- ANZAHL AUFRUFE AM TAG X IM DEZEMBER

def sort_per_day(csv_file, day):
    lines_for_day = []
    for i in range(len(csv_file)):
        line = csv_file.iloc[i]
        #print(line["Datum"][:2])
        if line["Datum"][:2] == str(day):
            #print(line["Datum"]) 
            lines_for_day.append(line)
        elif line["Datum"][:2] == "0" + str(day):
            #print(line["Datum"])
            lines_for_day.append(line)
    #print(f"Liste mit Linien für Tag {day}: {lines_for_day}")
    return lines_for_day
#sort_per_day(14)


# die werte für x und y -achse definieren
x_achse = [i for i in range(1, 25)]
#print(x_achse[0])

y_achse = []
for i in range(24):
    y_achse.append(len(sort_per_day(data, x_achse[i])))

print(f"X-Achse: {x_achse}")
print(f"Y-Achse: {y_achse}")


# diagramm zeichen mit x und y-achse
sns.barplot(x=x_achse, y=y_achse)

# Diagramm-Anpassungen
plt.title('Anzahl Aufrufe des Adventskalenders bis zum 24. Dezember')
plt.xlabel('Dezembertag')
plt.ylabel('Anzahl Aufrufe am Tag x')

# Diagramm anzeigen
plt.show()
```

```python
#-------------- ANZAHL AUFRUFE AM TAG X IM DEZEMBER UND JANUAR

# die werte für x und y -achse definieren
x_achse_dez = [i for i in range(1, 25)]
#print(x_achse[0])

x_achse_jan = [i for i in range(1, 12)]

x_achse_dez_jan = x_achse_dez + x_achse_jan

y_achse_dez_jan = []
for i in range(len(x_achse_dez_jan)):
    y_achse_dez_jan.append(len(sort_per_day(x_achse_dez_jan[i])))

x_achse_dez = [f"Dez-{i}" for i in range(1, 25)]
x_achse_jan = [f"Jan-{i}" for i in range(1, 12)]
x_achse_dez_jan = x_achse_dez + x_achse_jan

print(f"X-Achse: {x_achse_dez_jan}")
print(f"Y-Achse: {y_achse_dez_jan}")


# diagramm zeichen mit x und y-achse
sns.barplot(x=x_achse_dez_jan, y=y_achse_dez_jan)

# Diagramm-Anpassungen
plt.title('Anzahl Aufrufe des Adventskalenders vom 1.Dezember bis 11.Januar')
plt.xlabel('Datum')
plt.ylabel('Anzahl Aufrufe am Tag x')

# Diagramm anzeigen
plt.show()
```

```python
#-------------- ZAHLEN IM DEZEMEBER UND IM JANUAR

summe_aufrufe = sum(y_achse_dez_jan)
print(f"Summe aller Aufrufe gesamt: {summe_aufrufe}")

summe_aufrufe_dez = sum(y_achse)
print(f"Summe aller Aufrufe bis 24. Dezember: {summe_aufrufe_dez}")

durchschnittliche_aufrufe_dez = sum(y_achse)/24
print(f"Durchschnittliche tägliche Aufrufe bis 24. Dezember: {durchschnittliche_aufrufe_dez}")
```
