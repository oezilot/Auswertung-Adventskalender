```python
# try it out balkendiagramm
import matplotlib.pyplot as plt
import numpy as np

# erstellt die graphik
fig, ax = plt.subplots()

# daten die in der gtaphik dargestellt werden (2 datenarten?)
candidats = ('Marine Le Pen', 'Emmanuel Macron', 'François Fillon', 'Benoît Hamon', 'Jean-Luc Mélenchon')

voter = [21.7,23.9,20,6.3,19.2]

# sonstige einstellungen für die graphik (lände des diagramms und farbe)
y_pos = np.arange(len(candidats))
couleur = ['#000000', '#dbb243', '#2e42d3', '#e54fe3', '#f23434']

# beschriftungen und titel der graphik
plt.title('Election presidentielle 2017')
plt.ylabel('Vote en %')
plt.xlabel("Candidats")

# graphik generieren mit üarameter für das aussehen (alpha ist die deckkraft der farbe der graphen)
plt.bar(y_pos, voter, align='center', alpha=0.6, color=couleur)

# x seite der graphik ausrichten und länge definieren
ax.set_xticks(range(len(candidats)))
ax.set_xticklabels(candidats, rotation='vertical')

# ein png-file der graphik wird erstellt: (das muss vor dem plt.show command getan werden denn sonst erscheint eine blanke seite!)
plt.savefig('Diagramme/test.png')

# das ganze wird ausgeführt und eine graphi wird erstellt
plt.show()
```
